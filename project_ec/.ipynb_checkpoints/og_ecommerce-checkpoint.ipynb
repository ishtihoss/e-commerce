```python
# import necessary packages

# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# standard for xgboost
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb

# winsorize

from scipy.stats.mstats import winsorize

# date time

# more imports

from datetime import datetime
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import mean_squared_error, make_scorer
```

```python
# Load the data directly from GitHub
url = "https://raw.githubusercontent.com/ishtihoss/e-commerce/master/project_ec/og_ecommerce.csv"
data = pd.read_csv(url, encoding='Latin-1')

# Display the first 5 rows
data.head()
```

```python
# General info
print("Dataframe Summary:")
print(data.info())

# Detailed statistical summary for numerical columns
print("\nNumerical Columns Summary:")
print(data.describe())

# Summary of categorical columns
print("\nCategorical Columns Summary:")
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    print(f"\n{column} Column Summary:")
    print(data[column].value_counts())


```

```python
# Convert 'InvoiceDate' to datetime type
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Extract date, time, and day of the week and create new columns for each
data['Date'] = data['InvoiceDate'].dt.date
data['Time'] = data['InvoiceDate'].dt.time
data['DayOfWeek'] = data['InvoiceDate'].dt.day_name()

# Assign the new DataFrame to df
df = data.copy()

# Display the first few rows of df to check the new columns
df.head()
```

```python
# Create a new 'Revenue' column
df['Revenue'] = df['UnitPrice'] * df['Quantity']

# Determine the top 10 countries by total revenue
top_countries = df.groupby('Country')['Revenue'].sum().nlargest(10).index

# Filter the DataFrame to only include these countries
df_top_countries = df[df['Country'].isin(top_countries)]

# Group by country and day of the week, and sum the revenue
grouped = df_top_countries.groupby(['Country', 'DayOfWeek'])['Revenue'].sum().reset_index()

# Create a figure and an array of axes with a size based on the number of top countries
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(20, 10))

# Loop through each of the top countries and create a bar plot for each
for i, country in enumerate(top_countries):
    ax = plt.subplot(2, 5, i + 1)
    country_data = grouped[grouped['Country'] == country]
    ax.bar(country_data['DayOfWeek'], country_data['Revenue'])
    ax.set_title(country)
    plt.xticks(rotation=45)

# Improve layout
plt.tight_layout()
plt.show()
```

```python
# Create a 'Year' column
df['Year'] = df['InvoiceDate'].dt.year

# Group by 'Country' and 'Year', then sum 'Revenue'
grouped = df.groupby(['Country', 'Year'])['Revenue'].sum().reset_index()

# Sort the DataFrame by 'Revenue' in descending order
grouped = grouped.sort_values(by='Revenue', ascending=False)

# Create a figure and an axes object
fig, ax = plt.subplots(figsize=(10, 6))

# Loop through the unique years and plot the revenue for each year as a separate bar chart
for year in grouped['Year'].unique():
    year_data = grouped[grouped['Year'] == year]
    ax.bar(year_data['Country'], year_data['Revenue'], label=str(year))

# Improve the plot
ax.set_ylabel('Total Revenue')
ax.set_xlabel('Country')
plt.xticks(rotation=90)
ax.set_title('Total Revenue from Each Country per Year')
ax.legend()

plt.show()
```

```python
# Assuming 'InvoiceDate' is a datetime column, if not convert it first
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Now, let's extract the hour
df['Hour'] = df['InvoiceDate'].dt.hour

# Then proceed with the original operation
grouped = df.groupby(['Hour', 'DayOfWeek'])['Revenue'].median().reset_index()

# Pivot the DataFrame to have 'Hour' as index, 'DayOfWeek' as columns, and 'Revenue' as values
pivot = grouped.pivot(index='Hour', columns='DayOfWeek', values='Revenue')

# Plot the pivoted DataFrame
pivot.plot(kind='line', figsize=(10, 6))

# Improve the plot
plt.title('Mean Hourly Revenue by Day of Week')
plt.xlabel('Hour of Day')
plt.ylabel('Mean Revenue')
plt.xticks(range(0, 24))
plt.grid(True)

plt.show()
```

```python
# Filter UK only
df_uk = df[df['Country'] == 'United Kingdom']

# Group by 'Hour' and 'DayOfWeek', then find mean 'Revenue'
grouped_uk = df_uk.groupby(['Hour', 'DayOfWeek'])['Revenue'].mean().reset_index()

# Pivot the DataFrame to have 'Hour' as index, 'DayOfWeek' as columns, and 'Revenue' as values
pivot_uk = grouped_uk.pivot(index='Hour', columns='DayOfWeek', values='Revenue')

# Plot the pivoted DataFrame
pivot_uk.plot(kind='line', figsize=(10, 6))

# Improve the plot
plt.title('Mean Hourly Revenue by Day of Week in UK')
plt.xlabel('Hour of Day')
plt.ylabel('Mean Revenue')
plt.xticks(range(0, 24))
plt.grid(True)

plt.show()
```

```python
# Filter UK only
df_uk = df[df['Country'] == 'United Kingdom']
df_uk['Date'] = pd.to_datetime(df_uk['Date'])
```

```python
df_uk.dtypes
```

```python

# Convert 'DayOfWeek' to categorical variable
day_mapping = {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}
df_uk['DayOfWeek'] = df_uk['DayOfWeek'].map(day_mapping)

# Generate features based on the 'Quantity' column
# 'lag_{i}' will be the quantity i days ago
# 'ma_{i}' will be the moving average quantity over the past i days
for i in range(3, 13):
    print(f"Generating features for lag {i}")
    df_uk[f'lag_{i}'] = df_uk['Quantity'].shift(i)
    print(f"Generating features for moving average {i}")
    df_uk[f'ma_{i}'] = df_uk['Quantity'].rolling(window=i).mean()

print("Dropping NaNs")
    
# Drop the rows with missing values from the DataFrame
df_uk.dropna(inplace=True)

# Adjust the cutoff date
cutoff_date = df_uk['Date'].max() - pd.DateOffset(months=3)
train = df_uk[df_uk['Date'] <= cutoff_date]
test = df_uk[df_uk['Date'] > cutoff_date]

print("Number of rows in the adjusted test dataframe: ", test.shape[0])

# List of columns that XGBoost can't handle
drop_cols = ['InvoiceNo', 'StockCode', 'Description', 'Country', 'Time', 'InvoiceDate', 'Date', 'Revenue']

# Prepare the data for training
# We drop the target variable ('Quantity'), the 'Date' column (since we've handled time information in our features),
# and any other columns that XGBoost can't handle
X_train = train.drop(['Quantity'] + drop_cols, axis=1)
y_train = train['Quantity']
X_test = test.drop(['Quantity'] + drop_cols, axis=1)
y_test = test['Quantity']

print("Number of rows in the test dataframe: ", test.shape[0])
print("Number of elements in y_test: ", y_test.shape[0])


# Train the model
model = xgb.XGBRegressor()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate the root mean square error (RMSE) as a measure of the model's performance
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```

```python
xgb.plot_importance(model)
```

```python
from scipy.stats.mstats import winsorize
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import mean_squared_error, make_scorer

# Filter UK only
df_uk = df[df['Country'] == 'United Kingdom']
df_uk['Date'] = pd.to_datetime(df_uk['Date'])

# Convert 'DayOfWeek' to categorical variable
day_mapping = {'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7}
df_uk['DayOfWeek'] = df_uk['DayOfWeek'].map(day_mapping)

# Winsorize the 'Quantity' column to limit the effect of extreme values
df_uk['Quantity'] = winsorize(df_uk['Quantity'], limits=[0.05, 0.05])

# Generate features based on the 'Quantity' column
for i in range(3, 13):
    print(f"Generating features for lag {i}")
    df_uk[f'lag_{i}'] = df_uk['Quantity'].shift(i)
    print(f"Generating features for moving average {i}")
    df_uk[f'ma_{i}'] = df_uk['Quantity'].rolling(window=i).mean()

print("Dropping NaNs")
df_uk.dropna(inplace=True)

# Adjust the cutoff date
cutoff_date = df_uk['Date'].max() - pd.DateOffset(months=3)
train = df_uk[df_uk['Date'] <= cutoff_date]
test = df_uk[df_uk['Date'] > cutoff_date]

print("Number of rows in the adjusted test dataframe: ", test.shape[0])

# List of columns that XGBoost can't handle
drop_cols = ['InvoiceNo', 'StockCode', 'Description', 'Country', 'Time', 'InvoiceDate', 'Date', 'Revenue']

# Prepare the data for training
X_train = train.drop(['Quantity'] + drop_cols, axis=1)
y_train = train['Quantity']
X_test = test.drop(['Quantity'] + drop_cols, axis=1)
y_test = test['Quantity']

print("Number of rows in the test dataframe: ", test.shape[0])
print("Number of elements in y_test: ", y_test.shape[0])

# Define the model
model = xgb.XGBRegressor()

# Define the parameters for the grid search
parameters = {'n_estimators': [100, 200, 300], 
              'learning_rate': [0.01, 0.05, 0.1], 
              'max_depth': [3, 5, 7]}

# Define the scoring method
scorer = make_scorer(mean_squared_error, greater_is_better=False)

# Define the cross-validation method
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform the grid search
grid_obj = GridSearchCV(model, parameters, cv=cv, scoring=scorer)
grid_fit = grid_obj.fit(X_train, y_train)

# Get the best estimator
best_model = grid_fit.best_estimator_

# Train the best model
best_model.fit(X_train, y_train)

# Make predictions
y_pred = best_model.predict(X_test)

# Calculate the root mean square error (RMSE) as a measure of the model's performance
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```
